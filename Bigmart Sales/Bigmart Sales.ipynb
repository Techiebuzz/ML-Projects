```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
# importing dataset
train = pd.read_csv('D:\project\Grocery store\Train.csv')
test = pd.read_csv('D:\project\Grocery store\Test.csv')
X = train.iloc[:, 0:9].values
Y = train.iloc[:, 9].values
Y = train.iloc[0:,9].values
X_test = test.iloc[:, 0:9].values
```

```python
# Applying Algorithm
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X,Y)
```

```python
# Prediction
Y_pred = regressor.predict(X_test)
```
